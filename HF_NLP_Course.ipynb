**Interesting error**

The appearance of \ue002 (a private-use Unicode character) in your output typically indicates that the model is generating unknown tokens ([UNK]) or padding tokens that the tokenizer cannot map back to readable text. This is often caused by the *model failing to understand the prompt* or *getting "stuck" in a loop of invalid tokens*.

"looping" or generating gibberish (like those \ue002 Unicode characters) when it encounters tokens it doesn't quite know how to follow, or when the sampling settings aren't dialed in...

Code triggering the error:

```
text_generator = pipeline("text-generation")

text_generator("During Ramadan, I intend to ")
```



````python
text_generator = pipeline("text-generation")

text_generator("During Ramadan, I intend to ")
````

**Fixed by passing a more modern model & setting generation parameters to be more 'creative' (temperature)**

````python
# Workaround and cleaner code
from transformers import pipeline

# Using a more capable, modern, small model
text_generator = pipeline("text-generation", model="TinyLlama/TinyLlama-1.1B-Chat-v1.0")

prompt = "During Ramadan, I intend to"
output = text_generator(prompt, # same 'problematic' prompt
                        max_new_tokens=50,
                        do_sample=True,
                        temperature=0.7) # make output more human

print(output[0]['generated_text'])
````

**Load dataset**

````python
from datasets import load_dataset

imdb = load_dataset("imdb")
````

**Preprocess**

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert/distilbert-base-uncased")
````

````python
def preprocess_function(examples):
    return tokenizer(examples["text"], truncation=True)
````

````python
# batched speeds up the mapping function to process multiple elements simultaneously

tokenized_imdb = imdb.map(preprocess_function, batched=True)
````
