# drop() fxn modes:

**how:**

 any-> will drop the entire row if there is one NULL

  all-> will drop the entire row if ALL the feature values are NULL.



```
df_pyspark_DE.na.drop(how=all|any)
```


**thresh (threshold):**
define the minimum null values.

if null values > 2 then drop.

```
df_pyspark_DE.na.drop(thresh=2)
```

**subset:**

define a column. if there is null in this specific col then drop entry.



```
df_pyspark_DE.na.drop(subset=['<col name>'])
```



````python
# drop a column
# if the col name isn't define, the entire row with NULL will be dropped
df_pyspark_DE = df_pyspark_DE.na.drop(thresh=2)
df_pyspark_DE = df_pyspark_DE.na.drop(subset=['Name'])
````

````python
df_pyspark_DE.show()
````

# Filling values

1- Fill with a value:


```
df.na.fill('<value to fill with>', [col name list])
```

2- Fill using mean/mode/etc. , we use an Imputer:

```
from pyspark.ml.feature import Imputer

imputer = Imputer(
  inputCols=[col name list],
  outputCols=[]
).setStrategy("<strat>")
```


````python
from pyspark.ml.feature import Imputer

imputer = Imputer(
  inputCols=['age','Experience', 'Salary'],
  outputCols=["{}_imputed".format(c) for c in ['age','Experience', 'Salary']],
).setStrategy("mean")
````

````python
# add imputation cols to df

imputer.fit(df_pyspark_DE).transform(df_pyspark_DE).show()
````

# GroupBy

It is the prelilminary function used to filter the return data. Meaning, if I just groupby(x) and return:



> df_pyspark_agg.groupBy('Name').show()



SO we groupby the column we want our output data to be filtered by, then apply the aggregate function to do the filtering itself. Example below :)


````python
df_pyspark_agg.groupBy('Name').sum().show() # using their summed salaries, we grouped -summed- to find the max
````

````python
df_pyspark_agg.groupBy('Departments').sum().show() # what pays most
````

````python
df_pyspark_agg.groupBy('Name').min().show() # how many per dept
````

````python
df_pyspark_agg.agg({'Salary':'sum'}).show() # apply a function to a col directly
````
