Start a spark session to

```python
from pyspark.sql import SparkSession
```

```python
# creating a spark session
spark = SparkSession.builder.appName('Practise').getOrCreate()
```

```python
# Session information
spark
```

```python
# read file using the pyspark session

df_pyspark_og = spark.read.csv('test1.csv')
```

Below shows the data has been grabbed by spark using the session. We notice a discrepancy in the column names.

```python
df_pyspark_og.show()
```

The below check shows the datatypes were not grabbed from the csv file. To ensure our strings remain strings and ints remain ints we use the **inferSchema** parameter (as **True**) in the .csv function when pulling the file.

```python
df_pyspark_og.printSchema()
```

The below code ensures that while we read using spark, we retain the data's original headings also adding the inferSchema param.

Adding the **header** parameter set as True ensures consistency- then saving as a .csv.

```python
df_pyspark = spark.read.option('header', 'true').csv('test1.csv', inferSchema=True)
```

To see the df_pyspark datatype, we must ensure the variable is not assigned to a function (show.(), etc.) but just the csv (ending in .csv('<file path/<name>.csv>)').

If assigned to a function --> **NoneType**

Otherwise, as shown below:

```python
type(df_pyspark)
```

# Adding, dropping and renaming columns!

```python
# adding a column
df_pyspark= df_pyspark.withColumn('Experience after 10 years',
                      df_pyspark['Experience']+10 )
```

```python
df_pyspark.show()
```

```python
# dropping the column
df_pyspark = df_pyspark.drop('Experience after 10 years')
```

```python
df_pyspark.show()
```

```python
# renaming the column
df_pyspark = df_pyspark.withColumnRenamed('Name', 'New Name')
```
